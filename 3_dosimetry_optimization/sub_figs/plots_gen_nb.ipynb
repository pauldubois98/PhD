```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from matplotlib.colors import LinearSegmentedColormap
ncolors = 256
color_array = plt.get_cmap('hot_r')(range(ncolors))
color_array[:,-1] = np.linspace(0.0,1.0,ncolors)
map_object = LinearSegmentedColormap.from_list(name='hot_alpha',colors=color_array)
plt.colormaps.register(cmap=map_object)
```

```python
M = 5
N = 500
x = np.linspace(-M, M, N)
y = np.linspace(-M, M, N)
X, Y = np.meshgrid(x, y)
```

```python
M = 1.5
N = 500
x = np.linspace(-M, M, N)
y = np.linspace(-M, M, N)
X, Y = np.meshgrid(x, y)
Z = np.exp(-X**2 - Y**2)

plt.imshow(Z, cmap='hot_alpha', interpolation='bilinear', vmin=0, vmax=1)
plt.axis('off')
plt.savefig('hot_point.png', dpi=300, transparent=True, bbox_inches='tight')
plt.show()
```

```python
M = 3
N = 500
x = np.linspace(-M, M, N)
y = np.linspace(-M, M, N)
X, Y = np.meshgrid(x, y)
Z = np.exp(-0.5 * (0.1*np.clip(X**2 + Y**2-1, 0, None) + np.clip(X**2-1.5, 0, None) + np.clip(Y**2-4, 0, None) ) )
# Z += np.exp(-0.5*np.clip(X**2 + Y**2-1, 0, None))

plt.imshow(Z, cmap='hot_alpha', interpolation='bilinear', vmin=0, vmax=1.5)
plt.axis('off')
plt.savefig('hot_square.png', dpi=300, transparent=True, bbox_inches='tight')
plt.show()
```
